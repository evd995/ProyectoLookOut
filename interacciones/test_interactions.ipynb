```python
import numpy as np
import cv2
import pickle
import matplotlib.pyplot as plt
from shapely.geometry import Polygon
%matplotlib inline
```

```python
image = cv2.imread('test_data/class.png')

with open('test_data/cubes.pkl', 'rb') as file:
    cubes = pickle.load(file)
    
cubes
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
cube_1 = cubes[0]
cube_2 = cubes[1]
```

```python
sala = np.zeros(image.shape)
```

```python
from scipy.spatial import ConvexHull
hull_1 = ConvexHull(cube_1)
hull_2 = ConvexHull(cube_2)
```

```python
plt.fill(cube_1[hull_1.vertices,0], cube_1[hull_1.vertices,1], 'k', alpha=1)
```

```python
plt.fill(cube_2[hull_2.vertices,0], cube_2[hull_2.vertices,1], 'k', alpha=1)
```
